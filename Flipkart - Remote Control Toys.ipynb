```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.flipkart.com/toys/remote-control-toys/pr?sid=mgl,56m&otracker=nmenu_sub_Baby%20%26%20Kids_0_Remote%20Control%20Toys")
```

```python
print(response.status_code)
```

```python
print(response.content)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

### One Page

```python
cards = soup.find_all("div", attrs={"class":"_3liAhj"})

for card in cards:
    title = card.find("a", attrs={"class":"_2cLu-l"})
    price = card.find("div", attrs={"class":"_1vC4OE"})
    try:
        rating = card.find("div", attrs={"class":"hGSR34"}).text
    except:
        rating = None
    print("{} {} {}".format(title.text, price.text, rating))
```

### Mulltiple Pages

```python
home_url = "https://www.flipkart.com"
count = 0

while count < 2:
    cards = soup.find_all("div", attrs={"class":"_3liAhj"})
    
    for card in cards:
        title = card.find("a", attrs={"class":"_2cLu-l"})
        price = card.find("div", attrs={"class":"_1vC4OE"})
        try:
            rating = card.find("div", attrs={"class":"hGSR34"}).text
        except:
            rating = None
        print("{} {} {}".format(title.text, price.text, rating))
        
    next_url = soup.find("a", text="Next").get("href")
        
    response = requests.get(home_url + next_url)
    soup = BeautifulSoup(response.content, "html.parser")
        
    count += 1
    print(" ")
```

### One level Deep

```python
home_url = "https://www.flipkart.com"

cards = soup.find_all("div", attrs={"class":"_3liAhj"})


for card in cards:
    product_url = card.find("a", attrs={"class":"Zhf2z-"}).get("href")
    product_response = requests.get(home_url + product_url)
    product_soup = BeautifulSoup(product_response.content, "html.parser")
    title = product_soup.find("span", attrs={"class":"_35KyD6"}).text
    print(title)

    print(" ")
```
